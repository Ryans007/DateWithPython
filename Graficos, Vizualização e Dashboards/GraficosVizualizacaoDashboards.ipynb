Limpeza e Tratamento de Dados

```python
#Importação das Bibliotcas

import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as srn
```

```python
#Importação dos Dados

dados = pd.read_csv('dados.csv', sep= ';')
dados.head()
```

```python
srn.histplot(dados.PIB, kde=True, bins=50).set_title('Analise do PIB')
```

```python
srn.boxplot(dados.PIB, showfliers = False, orient= 'h').set_title('Analise do PIB')
```

```python
srn.boxplot(dados.PIB).set_title('Analise do PIB')
```

```python
srn.histplot(dados.VALOREMPENHO).set_title('Analise do VALOREMPENHO')
```

```python
plt.boxplot(dados.VALOREMPENHO)
plt.title('Analise do VALOREMPENHO')
```

```python
plt.figure(1)
plt.subplot(1,2,1)
srn.histplot(dados.PIB, kde=True, bins=30).set_title('Analise do PIB')
plt.subplot(1,2,2)
srn.histplot(dados.VALOREMPENHO, kde=True, bins=30).set_title('Analise do VALOREMPENHO')

```

```python
plt.scatter(x=dados.PIB, y=dados.VALOREMPENHO,marker='*', color='grey')
```

```python
srn.regplot(x=dados.VALOREMPENHO,y=dados.PIB, data= dados, x_jitter= 0.3, fit_reg=False)
```

```python
plt.plot(dados.PIB, dados.VALOREMPENHO)
```

```python
agrupado = dados.sort_values('PIB').head(10)
agrupado
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO', y='PIB', color= 'grey')
```

```python
agrupado = dados.sort_values('VALOREMPENHO').head(10)
agrupado
agrupado = agrupado.iloc[:,[1,3]]
agrupado
agrupado.plot.bar(x='MUNICIPIO', y='VALOREMPENHO',color='grey')
```
